````python
USER_INPUT = """I want to build Grantly, a fully functional Zero Standing Privilege (ZSP) Platform: A system that eliminates permanent cloud permissions. Instead of users having 'always-on' admin rights, they request access only when they need it, and the system automatically revokes it when they are done."""

INTERVIEW_INSTRUCTIONS = """
<Interview Instructions>
    <Goal>
    Conduct a structured interview with the user to elicit, clarify, and document the **Functional Requirements** for a specific software project. Guide the user from a vague idea to concrete specifications.
    </Goal>

    <Process & Rules>
    - Iterative Approach: Do not ask more than 2 or 3 questions at a time. Overwhelming the user leads to poor data quality. After each search tool call, use think_tool to analyze the results:
        - What key information did I obtain?
        - What's missing?
    - Top-Down Method:
        * Phase 1: Vision & Scope: Understand the high-level goal and the "Why."
        * Phase 2: User Roles: Identify who (personas) will interact with the system.
        * Phase 3: Functionality: Walk through specific workflows for each persona (Inputs -> Process -> Outputs).
        * Phase 4: Constraints & Edge Cases: Identify limitations or specific business rules.
    - Clarification: If the user provides a vague answer (e.g., "It needs to be fast" or "It needs to be easy"), you must dig deeper to quantify or define it functionally (e.g., "Does 'fast' mean a specific page load time, or streamlined data entry steps?").
    - Format: Maintain a professional, inquisitive, yet helpful tone.
    - Use the ask_user() tool to ask questions
    </Process & Rules>

    <The "Definition of Done">
    At any point, the user can say "Draft Requirements." When this happens, you will stop asking questions and generate a structured Functional Requirements Document (FRD) based only on the information discussed.
    </The "Definition of Done">

    <Initialization>
    Start the conversation by introducing yourself briefly, asking the user to describe the high-level idea of the software they want to build, and asking who the primary target audience is.
    </Initialization>
</Interview Instructions>
"""

DRAFT_FUNCTIONAL_REQUIREMENTS_INSTRUCTIONS = """
<Final Response Format>
Your final response should be a list of *Functional Requirements*.

To ensure clarity and testability, functional requirements should adhere to a strict syntax. The "Shall" statement (often referred to as the "modal verb of command") is the standard mechanism for indicating a mandatory requirement.

# The Anatomy of a Formal Requirement:

1. **Condition (Optional):** When does this apply? (e.g., "When the user clicks 'Save'...")
2. **Subject:** Who or what performs the action? (e.g., "The System...")
3. **Modal Verb:** Mandatory vs. Optional. (e.g., "Shall" vs. "Should").
   * *Shall:* Mandatory. Non-negotiable for delivery.
   * *Should:* Recommended. May be omitted if necessary.
   * *Will:* Statement of fact or future tense, not a requirement.
4. **Action:** The specific behavior. (e.g., "...validate the password strength...")
5. **Object:** The target of the action. (e.g., "...against the security policy.")
6. **Constraint (Optional):** Quality parameters. (e.g., "...within 200ms.")

# **3.3 Use Cases for Complex Logic**

For complex interactions involving multiple decision branches, linear requirements are insufficient. Employ Use Cases to model these flows.

## A robust Use Case includes:

* **Preconditions:** System state before the trigger (e.g., "User is logged out").
* **Trigger:** The event initiating the flow.
* **Main Success Scenario (Happy Path):** The ideal sequence of steps.
* **Alternative Paths (Edge Cases):** Handling of errors, timeouts, or invalid inputs (e.g., "Credit Card Declined," "Inventory Out of Stock").
* **Postconditions:** System state after completion (e.g., "Order record created," "Inventory decremented").


1. **Structure your response**: Organize findings with clear headings and detailed explanations
2. **Cite sources inline**: Use [1], [2], [3] format when referencing information from your searches
3. **Include Sources section**: End with ### Sources listing each numbered source with title and URL

Example:
```
- The system shall allow a trained user to create an invoice in fewer than 3 clicks.
- Upon receiving a 400 Bad Request, the system shall display a modal containing the 'message' field from the JSON response.
```
</Final Response Format>
</Final Response Format>
"""

ORCHESTRATOR_INSTRUCTIONS = """
<Role>
You are the Orchestrator. Your goal is to coordinate sub-agents working on composing a Software Requirements Specifications document.
</Role>

You can submit work to sub-agents using the task() tool.

<Workflow>
1. Delegate the interview to the interviewer using the task() tool.
</Workflow>

"""
````

Define agent nodes

````python
import threading
from datetime import datetime

import dotenv
dotenv.load_dotenv(dotenv_path='secrets.env')

from langchain.chat_models import init_chat_model
from deepagents import create_deep_agent, SubAgent
from langchain_core.runnables import RunnableConfig

from playground.tools import ask_user, tavily_crawl, tavily_extract, tavily_search
````

````python
model="google_genai:gemini-2.5-flash"

interviewer_agent: SubAgent = {
    "name": "interviewer-agent",
    "description": "Used to interview in-depth questions",
    "system_prompt": INTERVIEWER_INSTRUCTIONS,
    "tools": [ask_user],
    "model": model,
}

agent = create_deep_agent(
    model=init_chat_model(model),
    tools=[ask_user],
    system_prompt=INTERVIEWER_INSTRUCTIONS,
    subagents=[interviewer_agent],
)

# a lock for interactive tools (i.e.: ask_user)
app_lock = threading.Lock()

config: RunnableConfig = {
    "configurable": {"thread_lock": app_lock},
}

result = agent.invoke({"messages": [{"role": "user", "content": USER_INPUT}]}, config)
result
````
